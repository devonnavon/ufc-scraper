# Data Cleaning & Feature Engineering

```python
import pandas as pd
import numpy as np
import os
import json
from pathlib import Path

BASE_PATH = Path(os.getcwd())/'data'

with open(BASE_PATH/'fighters.json') as f: fighters = json.load(f)
with open(BASE_PATH/'fights.json') as f: fights = json.load(f)
    
df_fighters = pd.DataFrame(fighters).drop('fighter_id', axis=1).replace('--',np.NaN).replace('',np.NaN)
df_fights = pd.DataFrame(fights).drop(['fight_id', 'event_id'], axis=1).replace('--',np.NaN).replace('',np.NaN)
```
